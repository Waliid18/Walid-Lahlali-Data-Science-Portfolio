3. **Tirer 5 pourcents de l’échantillon sans remise.**

```python
pourcentage = 5
taille_echantillon = int((len(emissions1)*pourcentage/100))
echantillon_sans_remise = emissions1.sample(n= taille_echantillon, replace = False)
```

```python
echantillon_sans_remise
```

4. **Ne conserver que les 10 premières lignes et tirer aléatoirement dans celles-ci pour obtenir un DataFrame de 100 données.**

```python
les_dix_premieres_lignes = emissions1.head(10)
alt = les_dix_premieres_lignes.sample(n = 100 , replace = True)
alt
```

5. **Faire 100 tirages à partir des 6 premières lignes avec une probabilité de 1/2 pour la première observation et une probabilité uniforme pour les autres.**

```python
import numpy as np
Les_six_premières_lignes = emissions1.head(6)
```

```python
proba = [0.5] + [0.5/5]*5
```

```python
tir = np.random.choice(Les_six_premières_lignes.index,size = 100,p = proba)
df2 = emissions1.loc[tir]
```
