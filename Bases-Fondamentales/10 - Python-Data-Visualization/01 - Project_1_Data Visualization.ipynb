## 5. Histograms :

5.3 **The immigration distribution for France, USA, and China for years 1980 - 2013 :**

```python
YEARS = list(map(str,range(1980,2014)))
df_FUC = df_can.loc[['France','United States of America','China'],YEARS].transpose()
df_FUC.index = df_FUC.index.map(int)
count, bin_edges = np.histogram(df_FUC)
```

```python
df_FUC.plot(kind = 'hist' , figsize = (6,4) , color = ['coral', 'darkslateblue', 'mediumseagreen'] , alpha = 0.5, xticks = bin_edges , grid = True)

plt.xlabel('Number of Immigrants')
plt.ylabel('Number of Countries')
plt.title('Histogram of immigration distribution for France, USA, and China for years 1980 - 2013')

plt.show()
```
