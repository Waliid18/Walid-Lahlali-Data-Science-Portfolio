9. **Write a Pandas program to find out the alcohol consumption details in the year '1987' or '1989' from the world alcohol consumption dataset.**

```python
df[(df['Year']==1987) | (df['Year']==1989)]
```

10. **Write a Pandas program to find out the alcohol consumption details by the 'Americas' in the year '1985' from the world alcohol consumption dataset.**

```python
df[(df["Year"]==1985) & (df["WHO region"]== "Americas")]
```

11. **Write a Pandas program to find out the alcohol consumption details in the year '1986' where WHO region is 'Western Pacific' and country is 'VietNam' from the world alcohol consumption dataset.**

```python
df[(df["Year"]==1986) & (df["WHO region"]=="Western Pacific") & (df["Country"]=="Viet Nam")]
```

12. **Write a Pandas program to find out the alcohol consumption details in the year '1986' or '1989' where WHO region is 'Americas' from the world alcohol consumption dataset.**

```python
df[((df["Year"]==1986) | (df["Year"]==1989)) & (df["WHO region"]=="Americas")]
```
