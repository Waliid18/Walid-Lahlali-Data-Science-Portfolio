1. **Download the dataframe from the API call for Golden_State.**

**url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%205/Labs/Golden_State.pkl"**

```python
import pandas as pd
import requests

def download(url , file) :
    rep = requests.get(url)
    if rep.status_code == 200 :
        with open (file,'wb') as f :
            f.write(rep.content)

url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%205/Labs/Golden_State.pkl"
            
download(url , "Golden_State.pkl")
df = pd.read_pickle("Golden_State.pkl")
df.head()
```

5.**Plot out the PLUS MINUS column for the two dataframes home and away**
