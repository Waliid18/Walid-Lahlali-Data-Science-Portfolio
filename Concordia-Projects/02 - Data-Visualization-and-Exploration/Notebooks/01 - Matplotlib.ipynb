## Exercise 1: Recreate the Following Plots

### 1.1

Complete the code below and recreate the given plot.

Hint: To increase the size of the y-axis label, you can adjust the [`fontsize`](https://matplotlib.org/stable/users/explain/text/text_intro.html) parameter.

```python
# Data
x = np.linspace(0,10,500)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Plot here
# Create a new figure with a specified size

fig = plt.figure(figsize=(28, 12))

# Add three subplots to the figure

ax1 = fig.add_subplot(1, 3, 1)  # First subplot (1 row, 3 columns, first subplot)
ax2 = fig.add_subplot(1, 3, 2)  # Second subplot (1 row, 3 columns, second subplot)
ax3 = fig.add_subplot(1, 3, 3)  # Third subplot (1 row, 3 columns, third subplot)

# Plot the first dataset on the first subplot

ax1.plot(x, y1, linestyle='-', color='red', linewidth=5)  # Plot with a red line
ax1.set_xlabel('x', fontsize=45)  # Set the x-axis label with large font size
ax1.set_ylabel('sin(x)', fontsize=45)  # Set the y-axis label with large font size

# Plot the second dataset on the second subplot

ax2.plot(x, y2, linestyle='-', color='green', linewidth=5)  # Plot with a green line
ax2.set_xlabel('x', fontsize=45)  # Set the x-axis label with large font size
ax2.set_ylabel('cos(x)', fontsize=45)  # Set the y-axis label with large font size

# Plot the third dataset on the third subplot

ax3.plot(x, y3, linestyle='-', color='blue', linewidth=5)  # Plot with a blue line
ax3.set_xlabel('x', fontsize=45)  # Set the x-axis label with large font size
ax3.set_ylabel('tan(x)', fontsize=45)  # Set the y-axis label with large font size

# Adjust the spacing between the subplots

plt.subplots_adjust(wspace=0.4)

plt.show()
```

## Exercise 2

Canadians were asked to identify one of the following as official national food, here are the results:

| Food        | % of votes    |
|-------------|---------------|
| Bacon       | 35%           |
| Poutine     | 30%           |
| Salmon      | 17%           |
| Beavertail  | 8%            |
| Tourtiere   | 6%            |
| Doughnut    | 4%            |


Use the above (fake) data to recreate the following piecharts:

Hint: Refer to the `plt.pie()` method documentation to learn more about the parameters to customize a piechart.

### 2.1

```python
#2.1 Answer

# Creating a dictionary to hold food and percentage of votes data

data = {"Food" : ["Bacon","Poutine","Salmon","Beavertail","Tourtiere","Doughnut"],
"% of votes":[0.35,0.30,0.17,0.08,0.06,0.04]}

# Creating a pandas DataFrame from the dictionary

df = pd.DataFrame(data)
```

```python
# Plotting a pie chart using "% of votes" from the DataFrame and labeling each wedge with food names

plt.pie(df["% of votes"], labels = df.Food)

# Ensuring the pie chart is circular by setting the aspect ratio to equal

plt.axis('equal')

# Displaying the pie chart

plt.show()
```

### 2.2

```python
# 2.2 Answer 

# Plotting a pie chart using "% of votes" from the DataFrame, without labels on the wedges

plt.pie(df["% of votes"], labels = None)

# Ensuring the pie chart is circular by setting the aspect ratio to equal

plt.axis('equal')

# Adding a legend with labels from the DataFrame index and setting font size

plt.legend(labels = df.Food, fontsize = 9)

# Adding a title to the pie chart

plt.title('The result of the food survey')

# Displaying the pie chart

plt.show()
```

## Exercise 3

Load the penguins data set, located in the `Data` folder, and create the following graphs.

```python
# Load + Clean data first 

df = pd.read_csv('../Data/penguins.csv')
df = df.drop('Unnamed: 0', axis=1)
df = df.dropna()
df
```

```python
# 3 Answer
# Grouping the DataFrame 'df' by 'species' and calculating the mean of selected columns

df1 = df.groupby('species')[['bill_depth_mm', 'bill_length_mm', 'flipper_length_mm']].mean()

# Rounding the mean values to two decimal places for better readability

df1['bill_depth_mm'] = np.round(df1['bill_depth_mm'], 2)
df1['bill_length_mm'] = np.round(df1['bill_length_mm'], 2)
df1['flipper_length_mm'] = np.round(df1['flipper_length_mm'], 2)

# Displaying the resulting DataFrame 'df1'

df1
```

```python
ax = df1.plot(kind="bar", figsize=(10, 6))

# Annoter chaque barre avec sa valeur
for p in ax.patches:
    ax.annotate(str(p.get_height()), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points')

# Déplacer la légende au-dessus du graphique
plt.legend(loc='upper center', fancybox=True, shadow=True, ncol=3)

# Ajouter le titre centré
plt.title('Penguin attributes by species', fontsize=16, fontweight='bold')

# Ajouter l'étiquette de l'axe des y
plt.ylabel('Length (mm)', fontsize=14)

# Personnaliser les étiquettes de l'axe des x
plt.xticks(rotation=0, fontsize=12)

# Afficher le graphique
plt.tight_layout()
plt.show()
```
