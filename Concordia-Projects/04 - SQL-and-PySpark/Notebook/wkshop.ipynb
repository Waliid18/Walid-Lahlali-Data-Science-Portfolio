Using the same mtcars data, please execute the following:

**1.1** Select unique values in the year column and order them in acsending order. How many unique values are there?

**1.2** Create a column with vehicle age values

**1.3** If a vehicle is older than 45 years, then label it as "45+". Otherwise, label it as "<45". Create a new column that takes these two values

**1.4** How many vehicles are in our dataset that are under 45 and 45 and over?

```python
# Establish a connection to the 'mtcars.sqlite' database

connection = sqlite3.connect('mtcars.sqlite')
```

```python
# Execute the query to get the names of all tables in the database

pd.read_sql("SELECT name FROM sqlite_master WHERE type = 'table'",connection)
```

```python
# Select all rows and columns from the 'results' table

pd.read_sql_query('SELECT * FROM results',connection)
```

```python
# 1.1 - Unique values in the year column in acsending order

pd.read_sql_query(
                    """
                        SELECT DISTINCT year -- Select unique years
                        FROM results
                        ORDER BY year ASC -- Order the years in ascending order
                    """,connection)
```

```python
# 1.1 - How many unique values are there?

pd.read_sql_query(
                    """
                        SELECT COUNT(DISTINCT year) AS Numbre_of_unique_values -- Count the number of unique years
                        FROM results
                    """,connection)
```

```python
# 1.2 - Create a column with vehicle age values

pd.read_sql_query(
                    """
                        SELECT *,                           -- Select all columns
                        (2024-(1900+year)) AS vehicle_age   -- Calculate the age of the vehicle
                        FROM results
                    """,connection)
```

```python
# 1.3 - Column that takes these two values ("45+" and "<45")

pd.read_sql_query(
                    """
                        SELECT *,
                                CASE
                                    WHEN vehicle_age >= 45 THEN '45+'        -- Assign '45+' if vehicle_age is 45 or more
                                    ELSE '<45'                               -- Otherwise, assign '<45'
                                END Categorical_age                          -- Alias the result as 'Categorical_age'
                        FROM 
                            (
                                SELECT *,(2024-(1900+year)) AS vehicle_age   -- Calculate the vehicle age
                                FROM results 
                             )
                    """,connection)
```

```python
# 1.4 - Vehicle older than 45 years, with label "45+"

pd.read_sql_query(
                    """
                        SELECT name,vehicle_age                            -- Select columns 'name' and 'vehicle_age'
                        FROM                                               -- Subquery to calculate vehicle age
                            (                                               
                                SELECT *,(2024-(1900+year)) AS vehicle_age -- Calculate vehicle age
                                FROM results  
                            )
                        WHERE vehicle_age >= 45                            -- Filter for vehicles older than or equal to 45
                    """,connection)
```

```python
# 1.4 - Count of Vehicle older than 45 years, with label "45+" (METHOD 1)

pd.read_sql_query(
                  """
                     SELECT COUNT(*) AS Count_of_Vehicles_older_than_45 -- Count the number of vehicles older than 45 years
                     FROM 
                        (
                            SELECT *,(2024-(1900+year)) AS vehicle_age  -- Calculate vehicle age
                            FROM results  
                        )
                     WHERE vehicle_age >= 45                            -- Filter for vehicles older than or equal to 45
                 """,connection)
```

```python
# 1.4 - Vehicle youngest than 45 years, with label "<45"
pd.read_sql_query(
                    """
                        SELECT name,vehicle_age                            -- Select columns 'name' and 'vehicle_age'
                        FROM                                               -- Subquery to calculate vehicle age
                            (
                                SELECT *,(2024-(1900+year)) AS vehicle_age -- Calculate vehicle age
                                FROM results  
                            )
                        WHERE vehicle_age < 45                             -- Filter for vehicles younger than 45
                    """,connection)
```

```python
# 1.4 - Count of Vehicle youngest than 45 years, with label "<45" (METHOD 1)

pd.read_sql_query(
                    """
                        SELECT COUNT(*) AS Count_of_Vehicles_youngest_than_45 -- Count the number of vehicles younger than 45
                        FROM 
                            (
                            SELECT *,(2024-(1900+year)) AS vehicle_age        -- Calculate vehicle age
                            FROM results  
                            )
                        WHERE vehicle_age < 45                                -- Filter for vehicles younger than 45
                    """,connection)
```

```python
# 1.4 - How many vehicles are in our dataset that are under 45 and 45 and over? (METHOD 2)

pd.read_sql_query(
                    """
                        SELECT COUNT(*) AS Count_categorical_age,         -- Count the number of vehicles in each age category
                            CASE
                                WHEN vehicle_age >= 45 THEN '45+'         -- Assign '45+' if vehicle_age is 45 or more
                                ELSE '<45'                                -- Otherwise, assign '<45'
                            END Categorical_age                           -- Categorize the age into '45+' or '<45'
                        FROM 
                            (
                                SELECT *,(2024-(1900+year)) AS vehicle_age-- Calculate vehicle age
                                FROM results 
                             )
                        GROUP BY Categorical_age                          -- Group by the categorical age ('45+' and '<45')
                    """,connection)
```

```python
# 1.5 - All unique vehicle name that have 8 cylinders with horsepower that is greater than 125 but doesn't exceed 200

pd.read_sql_query(
                    """
                        SELECT DISTINCT name,cylinders,horsepower -- Select distinct combinations of name, cylinders, and horsepower
                        FROM results
                        WHERE cylinders = 8                       -- Where the number of cylinders is 8
                        AND horsepower BETWEEN 126 and 200        -- And the horsepower is between 126 and 200 (inclusive)
                    """,connection)
```

```python
# 2.2.1 Use a single query to see if your favorite song is in this dataset

# Define a variable with my favorite song's name

my_favorite_song = "Seven Nation Army"

# Select all data from the 'song_data' table where song_name matches my favorite song

pd.read_sql_query(f"""
                    SELECT *
                    FROM songs_table
                    WHERE song_name = "{my_favorite_song}"
                    """,connection1)
```

```python
# 2.2.2 How many unique popularity scores are there?

pd.read_sql_query(
                    """
                        SELECT COUNT(DISTINCT song_popularity) AS Count_of_unique_popularity_scores -- Count the number of unique song popularity scores
                        FROM songs_table
                    """,connection1)
```

```python
# 2.2.3 How many song names are in this dataset that contain 'radio edit' in their name?

pd.read_sql_query(
                    """
                        SELECT COUNT(song_name) AS 'Count radio edit' -- Count the number of occurrences of song names containing 'radio edit'
                        FROM songs_table
                        WHERE song_name LIKE '%radio edit%'           -- Where the song name contains 'radio edit'
                    """,connection1)
```

```python
# 2.2.4 - Write an SQL query to report songs with popularity of 99 and above. 

pd.read_sql_query(
                    """
                        SELECT  DISTINCT song_name,song_popularity -- Select distinct combinations of song_name and song_popularity
                        FROM songs_table
                        WHERE song_popularity >= 99                -- Where the song popularity is 99 or higher
                    """,connection1)
```

```python
# 2.2.4 - How many songs are there?

pd.read_sql_query(
                    """
                        SELECT COUNT (*) AS songs_score_above_99          -- Count the number of distinct songs with popularity 99 or higher
                        FROM
                            (
                                SELECT DISTINCT song_name,song_popularity -- Select distinct combinations of song_name and song_popularity
                                FROM songs_table
                                WHERE song_popularity >= 99               -- Where the song popularity is 99 or higher
                            )""",connection1)
```

```python
# 2.2.5 - Create a new column that contains song duration in the following format m:s. Example, the duration of Boulevard of Broken Dreams is 4:22

df = pd.read_sql_query('SELECT * FROM songs_table', connection1)

# Convert song duration from milliseconds to seconds

df['duration_sec'] = df['song_duration_ms'] / 1000

# Convert seconds to m:s format

df['duration_m_s'] = pd.to_datetime(df['duration_sec'], unit='s').dt.strftime('%M:%S')

# Update SQLite table with the new column

df.to_sql('songs_table_updated', con=connection1, index=False, if_exists='replace')

# Select data from 'songs_table_updated' to verify the new column is present

pd.read_sql_query('SELECT song_name, duration_m_s FROM songs_table_updated', connection1)
```

```python
# 2.2.6 The minimum, maximum, and average popularity scores

pd.read_sql_query(
                    """
                        SELECT MIN(song_popularity) AS min_popularity_scores, -- Select the minimum song popularity score
                        MAX (song_popularity) AS max_popularity_scores,       -- Select the maximum song popularity score
                        AVG(song_popularity) AS avg_popularity_scores         -- Select the average song popularity score
                        FROM songs_table
                    """,connection1)
```

```python
# 2.2.7 - The average popularity scores and count for songs that contain "love" 

pd.read_sql_query(
                    """
                        SELECT AVG(song_popularity) AS avg_love_songs      -- Select the average song popularity
                        FROM
                            (
                                SELECT DISTINCT song_name, song_popularity -- Select distinct combinations of song_name and song_popularity
                                FROM songs_table
                                WHERE song_name LIKE '%love%'              -- Where the song name contains 'love'  
                            )
                    """,connection1)
```

```python
# 2.2.7 - Average popularity scores and count for songs that do not contain 'love' in their name

pd.read_sql_query(
                    """
                        SELECT AVG(song_popularity) AS avg_no_love_songs      -- Select the average song popularity
                        FROM
                            (
                                SELECT DISTINCT song_name, song_popularity -- Select distinct combinations of song_name and song_popularity
                                FROM songs_table
                                WHERE song_name <> 'love'                  --  Where the song name is not equal to 'love'
                            )
                    """,connection1)
```
