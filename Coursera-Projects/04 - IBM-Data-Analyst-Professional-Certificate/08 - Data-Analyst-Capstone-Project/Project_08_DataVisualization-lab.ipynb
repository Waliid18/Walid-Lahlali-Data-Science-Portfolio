# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
converted_comp_data = pd.read_sql_query("SELECT ConvertedComp FROM master", conn)
plt.figure(figsize=(10, 6))
plt.hist(converted_comp_data['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
age_data = pd.read_sql_query("SELECT Age FROM master", conn)
plt.figure(figsize=(8, 6))
plt.boxplot(age_data['Age'], vert=False)
plt.title('Box plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
age_workweek_data = pd.read_sql_query("SELECT Age, WorkWeekHrs FROM master", conn)
plt.figure(figsize=(10, 6))
plt.scatter(age_workweek_data['Age'], age_workweek_data['WorkWeekHrs'], color='skyblue')
plt.title('Scatter plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
bubble_data = pd.read_sql_query("SELECT WorkWeekHrs, CodeRevHrs, Age FROM master", conn)
plt.figure(figsize=(10, 6))
plt.scatter(bubble_data['WorkWeekHrs'], bubble_data['CodeRevHrs'], s=bubble_data['Age'], alpha=0.5)
plt.title('Bubble plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

top_databases = pd.read_sql_query(QUERY, conn)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_databases['count'], labels=top_databases['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')
plt.show()
```

```python
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) as count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

top_lang = pd.read_sql_query(QUERY, conn)

print(top_lang)
```

```python
# Définir la requête SQL pour compter le nombre de répondants travaillant avec SQL
QUERY = """
SELECT COUNT(*) as count
FROM LanguageWorkedWith
WHERE LanguageWorkedWith LIKE '%SQL%'
"""
sql_users_count = pd.read_sql_query(QUERY, conn)

print("Respondents indicated that they currently work with 'SQL':", sql_users_count['count'].values[0])
```

```python
QUERY = """
SELECT COUNT(*) as count
FROM WebFrameWorkedWith
WHERE WebFrameWorkedWith = 'MySQL'
"""

mysql_only_count = pd.read_sql_query(QUERY, conn)

print("Respondents indicated that they work on 'MySQL' only:", mysql_only_count['count'].values[0])
```

```python
QUERY = """
SELECT COUNT(*) as count
FROM PlatformWorkedWith
WHERE PlatformWorkedWith = 'MySQL'
"""

# Exécuter la requête et récupérer les données
mysql_only_count = pd.read_sql_query(QUERY, conn)

# Afficher le nombre de répondants travaillant uniquement avec MySQL
print("Question 4")
print("How many respondents indicated that they work on 'MySQL' only:", mysql_only_count['count'].values[0])
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import matplotlib.pyplot as plt

# Define the SQL query to calculate median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
QUERY = """
SELECT Age, 
       AVG(WorkWeekHrs) as MedianWorkWeekHrs,
       AVG(CodeRevHrs) as MedianCodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
"""

# Execute the query and fetch the data
stacked_data = pd.read_sql_query(QUERY, conn)

# Plot the stacked chart
plt.figure(figsize=(10, 6))
plt.bar(stacked_data['Age'], stacked_data['MedianWorkWeekHrs'], label='Median WorkWeekHrs')
plt.bar(stacked_data['Age'], stacked_data['MedianCodeRevHrs'], label='Median CodeRevHrs', bottom=stacked_data['MedianWorkWeekHrs'])
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.legend()
plt.grid(True)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, 
       AVG(ConvertedComp) as MedianConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
"""

line_data = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 6))
plt.plot(line_data['Age'], line_data['MedianConvertedComp'], marker='o')
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

```python
# your code goes here
QUERY = """
SELECT Age, 
       AVG(ConvertedComp) as MedianConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30
GROUP BY Age
"""

line_data = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 6))
plt.plot(line_data['Age'], line_data['MedianConvertedComp'], marker='o')
plt.title('Median ConvertedComp for Ages 25-30')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
"""

bar_data = pd.read_sql_query(QUERY, conn)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(bar_data['MainBranch'], bar_data['Count'])
plt.title('Distribution of Respondents by MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(axis='x')
plt.show()
```
