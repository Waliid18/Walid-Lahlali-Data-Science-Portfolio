### 3. Feature Engineering

### 3.1 Goal Conversion Rate

Measures the **efficiency of goal scoring:** **`Goal¬†Conversion¬†Rate = Goals / Total¬†Attempts`**

```python
# Create a new column for Goal Conversion Rate
data['Goal Conversion Rate'] = data['Goal'] / data['Total Attempts']
data['Goal Conversion Rate'] = data['Goal Conversion Rate'].fillna(0)  # Handle division by zero
```

### 3.2 Pass Completion Rate

Measures **passing accuracy:** **`Pass¬†Completion¬†Rate = Passes¬†Completed / Passes`**

```python
# Create a new column for Pass Completion Rate
data['Pass Completion Rate'] = data['Passes Completed'] / data['Passes']
data['Pass Completion Rate'] = data['Pass Completion Rate'].fillna(0)  # Handle division by zero
```

### 1.1 Total Goals Scored by Each Team

```python
# Calculate total goals scored by each team
total_goals_per_team = data.groupby('Team')['Goal'].sum().sort_values(ascending=False)
total_goals_per_team
```

```python
# Create the path to save the figure
output_dir = "../Visualizations"
figure_path = os.path.join(output_dir, "Total Goals Scored by Each Team in FIFA World Cup 2022.png")

# Normalize the data for the colormap
norm = plt.Normalize(total_goals_per_team.min(), total_goals_per_team.max())

# Create a colormap object from the Purples palette
colormap = cm.Purples

# Apply colors to the bars using the colormap
colors = [colormap(norm(value)) for value in total_goals_per_team]

# Create the bar chart
plt.figure(figsize=(14, 8))  # Set the figure size
bars = total_goals_per_team.plot(kind='bar', color=colors, edgecolor='black')  # Plot with custom colors and black edges

# Add chart title and labels with bold formatting
plt.title('Total Goals Scored by Each Team in FIFA World Cup 2022', fontsize=16, fontweight='bold', color='#4B0082')  # Indigo
plt.xlabel('Teams', fontsize=12, fontweight='bold', color='#4B0082')
plt.ylabel('Total Goals', fontsize=12, fontweight='bold', color='#4B0082')

# Rotate the x-axis and y-axis tick labels for better readability
plt.xticks(rotation=45, fontsize=10, fontweight='bold', color='black')
plt.yticks(fontsize=10, fontweight='bold', color='black')

# Add the values above each bar in bold
for i, value in enumerate(total_goals_per_team):
    plt.text(i, value + 0.2, f'{value}', ha='center', va='bottom', fontsize=10, fontweight='bold', color='#4B0082')

# Adjust the layout to prevent clipping of labels
plt.tight_layout()

# Save the figure
plt.savefig(figure_path, dpi=300)  

# Show the plot
plt.show()
```

    + **1974:** **`Zaire` (now `DR Congo`)** became the **first Sub-Saharan African** nation to qualify for the World Cup.
 
    + **1978:** **`Tunisia`** was the **first African** team to **win** a World Cup match (3‚Äì1 vs. Mexico)

    + **1994:** **`Nigeria`** reached the **Round of 16**.
 
    + **1998:** **`Nigeria`** advanced to the **Round of 16** but lost 4‚Äì1 to Denmark. **`Morocco`** was eliminated in the **group stage**, narrowly missing out despite impressive performances. **`Cameroon`** and **`South Africa`** were both eliminated in the **group stage**.
      
    + **2002:** **`Senegal`** reached the **quarterfinals**, and **`South Africa`** also participated.
      
    + **2006:** **`Ghana`** reached the **Round of 16**.
      
    + **2010:** **`Ghana`** reached the **quarterfinals**, becoming the first African team to do so since Cameroon in 1990.
      
    + **2014:** **`Algeria`** reached the **Round of 16**.
      
    + **2018:** **No African** teams advanced past the group stage.
